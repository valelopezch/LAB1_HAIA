```python
from algorithm_core_test import CatanGA
import pandas as pd
```

```python
population_size = 10
generations = 10
mutation_prob = 1
crossover_prob = 1
selection_method = "tournament"
tournament_size = 4
mutation_sigma = 0.05
mutation_indpb = 0.1
num_games_per_individual = 100

ga = CatanGA(
    population_size=population_size, 
    generations=generations, 
    mutation_prob=mutation_prob, 
    crossover_prob=crossover_prob,
    selection_method=selection_method,
    tournament_size=tournament_size,
    mutation_sigma=mutation_sigma,
    mutation_indpb=mutation_indpb,
    num_games_per_individual=num_games_per_individual
)

best_distribution = ga.run()
```

```python
print("Best probability distribution and agent:", [float(x) for x in best_distribution[0]], best_distribution[1])
```

```python
df = pd.DataFrame(ga.logbook)
df
```
