```python
import numpy as np

def swap_probabilities_np(prob_array):

    sorted_indices = np.argsort(prob_array)
    swapped_probs = prob_array.copy()
    n = len(prob_array)
    
    for i in range(n // 2):
        low_idx, high_idx = sorted_indices[i], sorted_indices[n - i - 1]
        swapped_probs[low_idx], swapped_probs[high_idx] = swapped_probs[high_idx], swapped_probs[low_idx]

    return swapped_probs
```

```python
player = random.choices(AGENTS, weights=probs, k=1)[0]
print(player)
print(AGENTS.index(player))
aux = np.delete(probs, AGENTS.index(player))
aux_swap = swap_probabilities_np(aux)
print("Original:", aux)
print("Swapped:", aux_swap)
print("Sum of new probabilities:", aux_swap.sum())  
for _ in range(num_games_per_individual):
    opponents = random.choices([agent for agent in AGENTS if agent != player], weights=aux, k=3)
    print(opponents)
    players = [player] + opponents
    random.shuffle(players)
    print(players, '\n')
```
